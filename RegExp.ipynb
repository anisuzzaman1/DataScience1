```python
# import Lib
from pydataset import data
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
```

```python
data().head()
```

```python
# Matplot Init
%matplotlib notebook
```

```python
# Load Data
pima = data('Pima.tr')
```

```python
# Ploting Data
%matplotlib inline
pima.plot(kind='scatter', x='skin', y='bmi')
plt.show()
```

```python
# Test Train Split for Supervised Traninig
x_train,x_test, y_train, y_test = train_test_split(pima.skin, pima.bmi)
```

```python
# Test Train Visualization
plt.scatter(x_train, y_train, label="Training Data", color='r', alpha=.3)
plt.scatter(x_test, y_test, label="Test Data", color='g', alpha=.7)
plt.legend()
plt.title("Test and Train Data")
plt.show()
```

```python
# Create Linear Model and train it
LR = LinearRegression()
LR.fit(x_train.values.reshape(-1,1), y_train.values)
```

```python
# Use Model to predict on Test Data
prediction = LR.predict(x_test.values.reshape(-1,1))

# Plot prediction line against actual test data
plt.plot(x_test, prediction, label="Linear Regression", color='b')
plt.scatter(x_test,y_test, label="Actual Test Data", color='g', alpha=.7)
plt.legend()
plt.show()
```

```python
# Predict BMI of Woman with skin fold 50
LR.predict(np.array([[50]]))[0]
```

```python
# Score this model
LR.score(x_test.values.reshape(-1,1), y_test.values)
```
